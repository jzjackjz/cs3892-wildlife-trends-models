```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

We will now import the CSV datasets related to global temperature trends and wildlife populations (living planet index or LPI). We want to merge them into one dataframe to use for modeling and visualization.

```python
temperatures_df = pd.read_csv('../datasets/temperature-anomaly.csv')
lpi_df = pd.read_csv('../datasets/global-living-planet-index.csv')

temperatures_df = temperatures_df[["Year", "Global average temperature anomaly relative to 1961-1990"]]
lpi_df = lpi_df[["Year", "Living Planet Index"]]

merged_df = pd.merge(temperatures_df, lpi_df, on="Year", how="inner")

missing_values = merged_df.isnull().sum()
missing_values # check if there's missing data
```

```python
sns.set(style="whitegrid")

# Line plot for temperature anomalies over time
sns.lineplot(data=merged_df, x="Year", y="Global average temperature anomaly relative to 1961-1990")
plt.title('Temperature Anomalies Over Time')
plt.xlabel('Year')
plt.ylabel('Global Average Temperature Anomaly (°C)')
plt.show()

# Line plot for the Living Planet Index over time
sns.lineplot(data=merged_df, x="Year", y="Living Planet Index")
plt.title('Living Planet Index Over Time')
plt.xlabel('Year')
plt.ylabel('Living Planet Index')
plt.show()

# Scatter plot for temperature anomalies vs. Living Planet Index
sns.scatterplot(data=merged_df, x="Global average temperature anomaly relative to 1961-1990", y="Living Planet Index")
plt.title('Temperature Anomalies vs. Living Planet Index')
plt.xlabel('Global Average Temperature Anomaly (°C)')
plt.ylabel('Living Planet Index')
plt.show()
```
