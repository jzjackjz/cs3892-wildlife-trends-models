This Jupyter notebook will analyze the relationship between land use trends and global living planet index trends.

Data References:
- HYDE (2023) – with minor processing by Our World in Data. “Built-up area” [dataset]. PBL Netherlands Environmental Assessment Agency, “History Database of the Global Environment 3.3” [original data].
- World Wildlife Fund (WWF) and Zoological Society of London (2022) – processed by Our World in Data. “Central estimate – WWF and Zoological Society of London – Zoological Society of London” [dataset]. World Wildlife Fund (WWF) and Zoological Society of London, “Zoological Society of London” [original data].

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
```

```python
land_use_df = pd.read_csv('../datasets/land-use-over-the-long-term.csv')
lpi_df = pd.read_csv('../datasets/global-living-planet-index.csv')

land_use_df_adjusted = land_use_df[["Year", "Land use: Built-up area", "Land use: Grazingland", "Land use: Cropland"]].copy()
lpi_df_adjusted = lpi_df[["Year", "Living Planet Index"]].copy()

merged_df = pd.merge(land_use_df_adjusted, lpi_df_adjusted, on="Year", how="inner")
missing_values = merged_df.isnull().sum()
missing_values # check if there's missing data
```

```python
sns.set(style="whitegrid")

sns.lineplot(data=merged_df, x="Year", y="Land use: Built-up area")
plt.title('Land Use: Built-up Area Over Time')
plt.xlabel('Year')
plt.ylabel('Land Use: Built-up Area')
plt.show()

sns.lineplot(data=merged_df, x="Year", y="Living Planet Index")
plt.title('Living Planet Index Over Time')
plt.xlabel('Year')
plt.ylabel('Living Planet Index')
plt.show()

sns.scatterplot(data=merged_df, x="Land use: Cropland", y="Living Planet Index")
plt.title('Cropland Use vs. Living Planet Index')
plt.xlabel('Land use: Cropland')
plt.ylabel('Living Planet Index')
plt.show()

sns.scatterplot(data=merged_df, x="Land use: Built-up area", y="Living Planet Index")
plt.title('Built-up area Use vs. Living Planet Index')
plt.xlabel('Land use: Built-up area')
plt.ylabel('Living Planet Index')
plt.show()

sns.scatterplot(data=merged_df, x="Land use: Grazingland", y="Living Planet Index")
plt.title('Grazingland Use vs. Living Planet Index')
plt.xlabel('Land use: Grazingland')
plt.ylabel('Living Planet Index')
plt.show()
```
