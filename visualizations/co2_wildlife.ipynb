```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
co2_df = pd.read_csv('../datasets/co2-long-term-concentration.csv')
lpi_df = pd.read_csv('../datasets/global-living-planet-index.csv')

co2_df_adjusted = co2_df[["Year", "Long-run CO₂ concentration"]].copy()
lpi_df_adjusted = lpi_df[["Year", "Living Planet Index"]].copy()

merged_df = pd.merge(co2_df_adjusted, lpi_df_adjusted, on="Year", how="inner")
missing_values = merged_df.isnull().sum()
missing_values # check if there's missing data
```

```python
sns.set(style="whitegrid")

# Line plot for temperature anomalies over time
sns.lineplot(data=merged_df, x="Year", y="Long-run CO₂ concentration")
plt.title('CO2 Concentration Over Time')
plt.xlabel('Year')
plt.ylabel('Long-run CO2 concentration (ppmv)')
plt.show()

# Line plot for the Living Planet Index over time
sns.lineplot(data=merged_df, x="Year", y="Living Planet Index")
plt.title('Living Planet Index Over Time')
plt.xlabel('Year')
plt.ylabel('Living Planet Index')
plt.show()

# Scatter plot for temperature anomalies vs. Living Planet Index
sns.scatterplot(data=merged_df, x="Long-run CO₂ concentration", y="Living Planet Index")
plt.title('CO2 Concentration vs. Living Planet Index')
plt.xlabel('Long-run CO2 concentration (ppmv)')
plt.ylabel('Living Planet Index')
plt.show()
```
