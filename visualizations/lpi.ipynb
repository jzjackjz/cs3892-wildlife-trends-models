```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
```

```python
lpi_df = pd.read_csv('../models/predicted_lpi_2024_2124.csv')

plt.plot(lpi_df['Year'], lpi_df['Predicted_LPI'], marker='o', linestyle='-', color='blue')  # Creates the line plot

# Customizing the plot
plt.title('Predicted Logistics Performance Index (LPI) Over Time')  # Adds a title
plt.xlabel('Year')  # Adds a label to the x-axis
plt.ylabel('Predicted LPI')  # Adds a label to the y-axis
plt.grid(True)  # Adds a grid
plt.xticks(rotation=45)  # Rotates the x-axis labels for better readability

# Display the plot
plt.show()
```
